```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load CSV data
def load_csv(filename):
    data_list = []
    with open(filename, 'r', encoding='utf-8-sig') as f:
        for line in f:
            row = [float(val.replace(';', '.')) for val in line.split(',')]
            data_list.append(row)
    return np.array(data_list)

# Visualize the colorful CSV image
def visualize_colorful_csv(csv_filename):
    # Load the CSV data
    csv_data = load_csv(csv_filename)
    
    # Normalize the data to [0, 255]
    normalized_data = (255 * (csv_data - csv_data.min()) / (csv_data.max() - csv_data.min())).astype(np.uint8)
    
    # Apply the jet colormap
    colorful_image = cm.jet(normalized_data / 255.0)
    
    # Display the image
    plt.figure(figsize=(10, 5))
    plt.imshow(colorful_image)
    plt.title("Colorful CSV Image")
    plt.axis('off')
    plt.show()

# Run the visualization function
visualize_colorful_csv('scone_1.csv')
```
